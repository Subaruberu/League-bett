Determinando as variáveis X e Y

```python
X = dados.drop('WR',axis=1).values
Y = dados['WR'].values
```

Aplicando regressão RIDGE

```python
from sklearn.linear_model import Ridge
```

Aplicando inicialmente com coeficiente alpha=0.5

```python
modelo1 = Ridge(alpha=0.5)
modelo1.fit(X, Y)
```

```python
score = modelo1.score(X, Y)
```

```python
score
```

Determinando o score com diferentes lambdas

```python
# Definindo os valores de lambda
lambdas = [0.00001, 0.0001, 0.001, 0.01, 0.1, 0.5, 1, 2, 5]

# Lista para armazenar os scores
scores = []

# Calculando os scores para cada valor de lambda
for i in range(len(lambdas)):
    modelo = Ridge(alpha=lambdas[i])
    modelo.fit(X, Y)
    score = modelo.score(X, Y)
    if score is not None:  # Verificando se o score foi calculado com sucesso
        scores.append(score)
    else:
        print(f"Erro ao calcular score para lambda = {lambdas[i]}")

# Criando um DataFrame com os scores e lambdas
df = pd.DataFrame({"Lambda": lambdas[:len(scores)], "Score": scores})

# Ordenando o DataFrame pelo valor de "Lambda" em ordem decrescente
df = df.sort_values(by="Score", ascending=False)

# Reiniciando o índice do DataFrame
df = df.reset_index(drop=True)

# Imprimindo o DataFrame ordenado
print(df)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Carregar os dados do arquivo Excel
file_path = 'LCK Data.xlsx'
Clean_Data = pd.read_excel(file_path)

Clean_Data = Clean_Data.drop('GD', axis=1)
Clean_Data = Clean_Data.drop('Region', axis=1)
Clean_Data = Clean_Data.drop('Season', axis=1)
Clean_Data = Clean_Data.drop('Games', axis=1)
Clean_Data = Clean_Data.drop('WR', axis=1)

# Lista de coeficientes (substitua pelos coeficientes reais que você forneceu)
coeficientes = variaveis_list

# Verificar o número de colunas de dados (exceto a coluna 'Name')
numeric_columns = Clean_Data.select_dtypes(include=[float, int]).columns
num_columns = len(numeric_columns)

# Ajustar a lista de coeficientes se necessário
if len(coeficientes) < num_columns:
    coeficientes += [0.0] * (num_columns - len(coeficientes))  # Preencher com zeros se houver menos coeficientes
elif len(coeficientes) > num_columns:
    coeficientes = coeficientes[:num_columns]  # Truncar se houver mais coeficientes

# Criar a nova coluna resultante da multiplicação
Clean_Data['Result'] = 25  # Inicializa a coluna com zeros

# Multiplicar cada coluna numérica pelos coeficientes correspondentes e somar na nova coluna
for i, column in enumerate(numeric_columns):
    Clean_Data['Result'] += (Clean_Data[column] * coeficientes[i])
    print(f'coluna {column} valor da coluna {Clean_Data[column]} / valor do coeficiente {coeficientes[i] }')

# Exibir o DataFrame resultante com a nova coluna
print(Clean_Data)
```

```python
import pandas as pd

# Carregar o arquivo Excel
file_path = 'LCK Data.xlsx'
df = pd.read_excel(file_path, sheet_name='Sheet1')

# Função para encontrar o time pelo nome
def find_team_by_name(team_name):
    # Busca o time pelo nome
    team_info = df[df['Name'].str.contains(team_name, case=False, na=False)]

    # Verifica se o time foi encontrado
    if team_info.empty:
        return f"Time '{team_name}' não encontrado."
    else:
        return team_info

# Lista de nomes dos times
team_names = ['Dplus KIA', 'DRX', 'FearX', 'Gen.G eSports', 'Hanwha Life eSports',
              'KT Rolster', 'Kwangdong Freecs', 'Nongshim RedForce', 'OK BRION', 'T1']

# Buscar informações de cada time e imprimir os resultados
for team_name in team_names:
    print(f"Informações para {team_name}:")
    print(find_team_by_name(team_name))
    print("\n" + "="*50 + "\n")
```

**Função para gerar uma Comparação de coeficientes**


```python
def calcular_winrate(df, time_a, time_b):
    # Obter os resultados dos dois times
    resultado_a = df.loc[df['Name'] == time_a, 'Result'].values[0]
    resultado_b = df.loc[df['Name'] == time_b, 'Result'].values[0]

    # Calcular o número necessário
    numero_necessario = 100 / (resultado_a + resultado_b)

    # Calcular os produtos
    produto_a = resultado_a * numero_necessario
    produto_b = resultado_b * numero_necessario

    # Exibir os resultados
    print(f'A winrate do time {time_a} é {produto_a}')
    print(f'A winrate do time {time_b} é {produto_b}')

    return numero_necessario, produto_a, produto_b

time_a = "Gen.G eSports"
time_b = "DRX"

numero, winrate_a, winrate_b = calcular_winrate(df, time_a, time_b)
```
